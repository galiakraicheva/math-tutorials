## Definition: 
A function is a law that takes an input and produces a single output. 

## One-to-one and many-to-one functions
- **many-to-one rules:** a single input can produce more than one output. It is not a function. Example: $ \sqrt{x} $
- **many-to-many rules:** a single input can produce more ouptputs and a single output can result from more than one input. It is not a function. Example: a circle $ x^2 + y^2 = r^2 $; other examples are: an elipse $ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1 $; a hyperbola $ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1 $; other equations: $ \sin(x) + \cos(y) = 1 $ and $ x^3 + y^3 = 1 $
- **one-to-many rules/functions:** a single input produces more than one output. It is a function. Example: $ y = ax^2 + bx + c $
- **one-to-one rules/functions:** s single input produces a single output. It is a function. Example: linear function, exponential function, logarithmic function.

## Inverse functions
Only one-to-one functions have a inverse function: a function that reverses the process of the original function. It takes the output of the original function and produces the input of the original function. 

If the original function is: $ f(x) $, its inverse function is written like this: $ f^{-1}(x) $.

If $ f(a) = b $ , then $ f^{-1}(b) = a $.

To find the inverse of a function, change the places of x and y in the original function and solve for y. 

## Continuous and discontinuous functions
A function is continuous if you can draw it without lifting your pencil from beginning to end. A function is discontinuous if you have to lift your pencil or if there are any holes where the function is not defined or it changes the rule.

```python
# Drawing a continuous and a discontinuous function
import numpy as np
import matplotlib.pyplot as plt

def continuous_function(x):
    return np.sin(x)  # Example: Sine function

def discontinuous_function(x):
    return np.where(x != 0, 1 / x, 0)  # Example: 1/x with a "hole" at x=0

# Create x values for both functions
x_continuous = np.linspace(-2 * np.pi, 2 * np.pi, 500)
x_discontinuous = np.linspace(-10, 10, 500)
x_discontinuous = x_discontinuous[x_discontinuous != 0]

# Plot of the continuous function
plt.figure(figsize=(8, 6))
plt.plot(x_continuous, continuous_function(x_continuous), label="Continuous: $y = \sin(x)$")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Graph of a Continuous Function", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Plot of the discontinuous function
plt.figure(figsize=(8, 6))
plt.plot(x_discontinuous, discontinuous_function(x_discontinuous), label="Discontinuous: $y = 1/x$")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Graph of a Discontinuous Function", fontsize=16)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

A common scenario for a discontinuous function is also the rule to change in the range in which the function is defined. For example: 
$$
f(x) = \begin{cases} 
3 & \text{if } x < 0 \\ 
x^2 & \text{if } x \geq 0 
\end{cases}
$$

```python
# Define the range of x
x1 = np.linspace(-5, 0, 400)  # Values for x < 0
x2 = np.linspace(0, 5, 400)   # Values for x >= 0

# Define the function values
y1 = 3 * np.ones_like(x1)  # f(x) = 3 for x < 0
y2 = x2**2                 # f(x) = x^2 for x >= 0

# Plot the function
plt.plot(x1, y1, label='f(x) = 3 (x < 0)') # Blue color is by default
plt.plot(x2, y2, label='f(x) = x² (x ≥ 0)', color='orange')

# Add labels, title, and legend
plt.title('Graph of the Piecewise Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=1.5)  # x-axis
plt.axvline(0, color='black', linewidth=1.5)  # y-axis
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()

# Show the plot
plt.show()
```

**What value does y approach when we approach 0 from the left?** 
Obvioulsy, y approaches 0. We write that like this: 
$$\lim_{x \to 0^+} f(x) = 0$$

**What value does y approach when we approach 0 from the right?**
Obviously, y approaches 3. We write that like this: 
$$\lim_{x \to 0^-} f(x) = 3$$

So a function is continuous if and only if for all a belonging in the range of the function:
$$
\lim_{x \to a^+} f(x) = \lim_{x \to a^-} f(x) = f(a)
$$

## Periodic functions

**Periodic Function:** a function that has a definite pattern repeated at regular intervals. (HELM Workbook 2, p. 29)

**Period of a function:** the interval over which the repetition takes place (HELM Workbook 2, p.29). It is usually denoted by T. 

If a function is periodic: $ f(x) = f(x + T) $

## Odd and even functions

**An even function:** a function that is symmetrical about the y-axis. For such functions: $ f(x) = f(-x) $

**An odd function:** a function with rotational symmetry about the origin. This means that each curve, when rotated at $ 180^\circ $, transforms into itself. For such functions: $ f(x) = - f(-x) $ 

For example: $ f(x) = x^3 $

```python
x = np.linspace(-2, 2, 200)

y = x**3

# Plot the function
plt.plot(x, y, label='y = x³')

plt.axis('equal')

plt.title('Graph of the Odd Function y = x³')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1.5)  # x-axis
plt.axvline(0, color='black', linewidth=1.5)  # y-axis
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()

plt.show()
```

We see that if we rotate the function's graph with $ 180^\circ ^, it is still itself. 

## The gradient of a line through 2 points

gradient (slope) of the line: $ \Delta = \frac{y_2 - y_1}{x_2 - x_1} $

## Distance between two points

The distance between any two points $ A(x_1, y_1) $ and $ B(x_2, y_2) $: $ AB = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2} $

## Circle at the origin of a coordinate system

The equation of a circle with its origin at the origin of the coordinate system: $\sqrt{x^2 + y^2} = R$, or more usually, $x^2 + y^2 = R^2$

This follows from the equation of the distance between two points. 

Generally, the equation of the circle with a centre $O(x_0, y_0)$ and a radius $R$ is: $(x - x_0)^2 + (y - y_0)^2 = R^2 \tag{2})$, which is said to be the equation of the circle with a centre $O(x_0, y_0)$ and a radius $R$.

$ R = \sqrt{(x - x_0)^2 + (y - y_0)^2} $ 

## Points inside and outside of a circle
If for a point $ A(x, y)$: <br>
$ x^2 + y^2 = R^2 $, the point is on the circle <br>
$ x^2 + y^2 > R^2 $, the point is outside the circle <br>
$ x^2 + y^2 < R^2 $, the point is inside the circle. 

## Annular regions
**Annular region:** the area between two concentric circles (circles with the same origin) 
Example: 2 circles with centers at the origin and radii 1 and 2 <br>
The set of points that lies in the annular region satisfies both equations: <br>
$ x^2 + y^2 > 1 $ (these are the points that are outside of the smaller circle and <br>
$ x^2 + y^2 < 4 $ (these are teh points that are inside the bigger circle. 

```python
#Set x ranges for each circle
x1 = np.linspace(-1, 1, 500)  # For the smaller circle (radius = 1)
x2 = np.linspace(-2, 2, 500)  # For the larger circle (radius = 2)

# Circle 1 (radius = 1)
y1_positive = np.sqrt(1 - x1**2)
y1_negative = -np.sqrt(1 - x1**2)

# Circle 2 (radius = 2)
y2_positive = np.sqrt(4 - x2**2)
y2_negative = -np.sqrt(4 - x2**2)

# Plot the first circle
plt.plot(x1, y1_positive, label='Radius = 1', color='blue')
plt.plot(x1, y1_negative, color='blue')

# Plot the second circle
plt.plot(x2, y2_positive, label='Radius = 2', color='orange')
plt.plot(x2, y2_negative, color='orange')

plt.axis('equal')  # Equal scaling for both axes

plt.title('Two Circles with Origin at (0, 0)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=1.5)  # x-axis
plt.axvline(0, color='black', linewidth=1.5)  # y-axis
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()

plt.show()
```

The area that satisfies this equation is between the orange and the blue circle. 

## Asymptotes 

**An asymptote:** a line on the graph that the function reaches closer and closer to but never crosses or touches the line. 
**A horizontal asymptote:** 

```python
# Setting x ranges, splitting at x = -1 to avoid the visual connection
x_left = np.linspace(-3, -1, 250, endpoint=False)  # Values strictly less than -1
x_right = np.linspace(-1, 3, 250)[1:]  # Values strictly greater than -1

# Defining the function as two functions
y_left = (2 * x_left + 1) / (x_left + 1)
y_right = (2 * x_right + 1) / (x_right + 1)

# Plotting the two functions
plt.plot(x_left, y_left, label=r'$f(x) = \frac{2x + 1}{x + 1}$', color='blue')
plt.plot(x_right, y_right, color='blue') # Need the same color, otherwise the two parts will be colored differently

# Drawing the asymptotes
plt.axhline(y=2, color='red', linestyle='--', label='Asymptote: y = 2')
plt.axvline(x=-1, color='purple', linestyle='--', label='Asymptote: x = -1')

# Setting axis limits to emphasize asymptotes
plt.ylim(-10, 10)  # Restrictting the y-axis range to focus on behavior near the asymptote
plt.xlim(-3, 3)    # No need to change this

# Labels, title and legend
plt.title('Graph of a Function with a Horizontal and Vertical Asymptote')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=1.5)  # x-axis
plt.axvline(0, color='black', linewidth=1.5)  # y-axis
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()

plt.show()
```

There may be oblique/slant asymptotes too. These are lines that the function's graph approaches but never reaches. One such function with an oblique asymptote is $ f(x) = \frac{x^2 + x + 1}{x} $. It is present because the degree in the numerator is 1 degree higher than the degree in the denominator. 

If we want to represent the oblique asymptote on the graph we need to calculate it and add it manually. To find the asymptote, we perform polinomial long division: 

$$
\frac{x^2 + x + 1}{x} = x + 1 + \frac{1}{x}
$$

Therefore, as  $ x \to \pm\infty $, the term  $\frac{1}{x} \to 0 $, so the slant asymptote is  $ y = x + 1 $.


```python
# Set the x range, excluding x = 0 to avoid division by zero
x_left = np.linspace(-10, -0.1, 500)  # Values strictly less than 0
x_right = np.linspace(0.1, 10, 500)   # Values strictly greater than 0

# Define the function
f_left = (x_left**2 + x_left + 1) / x_left
f_right = (x_right**2 + x_right + 1) / x_right

# Define the slant asymptote: y = x + 1
y_slant_left = x_left + 1
y_slant_right = x_right + 1

# Plot the function
plt.plot(x_left, f_left, label=r'$f(x) = \frac{x^2 + x + 1}{x}$', color='blue')
plt.plot(x_right, f_right, color='blue')

# Plot the slant asymptote
plt.plot(x_left, y_slant_left, linestyle='--', color='red', label='Asymptote: y = x + 1')
plt.plot(x_right, y_slant_right, linestyle='--', color='red')

# Add vertical asymptote at x = 0
plt.axvline(0, color='purple', linestyle='--', label='Asymptote: x = 0')

# Set axis limits
plt.ylim(-10, 10)  # Focus on the interesting range of y-values
plt.xlim(-10, 10)  # Cover the range of x-values

# Labels, title and legend
plt.title('Graph of $f(x) = \\frac{x^2 + x + 1}{x}$ with Asymptotes')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=1.5)  # x-axis
plt.axvline(0, color='black', linewidth=0.5)  # y-axis, smaller linewidth to make the asymptote more visible
plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend()

plt.show()
```
